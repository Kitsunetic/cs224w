```python
import networkx as nx
```

```python
g1 = nx.Graph()
```

```python
g1.add_node("a")
g1.add_node(1)
g1.add_node(2)
g1.add_node(4)

g1.add_nodes_from([11, 22])
```

```python
nx.draw(g1, with_labels=True, font_weight="bold")
```

```python
g1.remove_node(4)
```

```python
nx.draw(g1, with_labels=True, font_weight="bold")
```

```python
g1.add_edge(1, "a")
g1.add_edge(1, 2)
g1.add_edge(1, 22)
```

```python
nx.draw(g1, with_labels=True, font_weight="bold")
```

```python
g1.remove_node(22)
```

```python
nx.draw(g1, with_labels=True, font_weight="bold")
```
